# Function for generating an image masking for the segmentated detections

```python
##Modify the path to your photos and the json result path 
images_path = "/Users/josephinebocquet/Documents/Data_Josephine/GOPRO_22-09-12/Groupe2" 
json_path = "/Users/josephinebocquet/Documents/Darwin/result/"

def tree_masking(image_name): 
    img = os.path.join(images_path, image_name)
    json_file = os.path.join(json_path, image_name+".json")

    # Load an image
    image = cv2.imread(img)
    
    # Convert color from BGR to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Convert the image data to float format
    image = image.astype(float)

    with open(json_file, "r") as f:
        result = json.load(f)
    
    bin_img = image.copy()

    # Create a blank mask with the same size as the image

    height, width, _ = image.shape
    
    x_coords = []
    y_coords = []   
    for item in result["result"]:
        x_coords = []
        y_coords = [] 
        
        polygons = item["polygon"]["path"]
        
        
        for point in polygons:
            x_coords.append(int(point['x']))
            y_coords.append(int(point['y']))

            coords = np.column_stack((x_coords, y_coords))

            cv2.fillPoly(bin_img, [coords], color=(0, 0, 0))

    # Apply the mask to the image
    image[bin_img == (0, 0, 0)] = np.nan

    return image
```

```python
import cv2
import piexif
import cv2

def extract_metadata(images_path):
    image = cv2.imread(images_path, cv2.IMREAD_ANYCOLOR)
    metadata = image._getexif()
    return metadata

# Usage
metadata = extract_metadata(images_path)


def import_metadata(source_file, target_file, metadata):
    image = cv2.imread(source_file, cv2.IMREAD_UNCHANGED)

    # Extract the original metadata from the source file
    exif_dict = piexif.load(source_file)

    # Update the metadata with the new values
    exif_dict.update(metadata)

    # Convert the updated metadata to bytes
    exif_bytes = piexif.dump(exif_dict)

    # Save the image with the updated metadata
    cv2.imwrite(target_file, image, [cv2.IMWRITE_JPEG_QUALITY, 100, cv2.IMWRITE_JPEG_OPTIMIZE, True, cv2.IMWRITE_JPEG2000_COMPRESSION_X1000, 1000], exif=exif_bytes)

# Usage

import_metadata(images_path, result_path, metadata)
```
