```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import json
```

```python
image_path = "/Users/josephinebocquet/Documents/Darwin/Documents/Darwin/fub-remote-sensing/pine_forest_plot1/images/GOPR0928.JPG"
image = plt.imread(image_path)
```

```python
json_path = "/Users/josephinebocquet/Documents/Darwin/result/response.json"
with open(json_path, "r") as f:
    result = json.load(f)
```

```python
fig, ax = plt.subplots()
ax.imshow(image)

for item in result["result"]:
    label = item["label"]
    bbox = item["bounding_box"]
    polygon = item["polygon"]["path"]

    # Create a polygon patch for the polygon
    polygon_pts = [(pt["x"], pt["y"]) for pt in polygon]
    poly = patches.Polygon(polygon_pts, linewidth=1, edgecolor="b", facecolor="none")
    ax.add_patch(poly)

    # Add label text
    ax.text(bbox["x"], bbox["y"], label, color="r", fontsize=8)

# Adjust the plot limits
plt.xlim(0, image.shape[1])
plt.ylim(image.shape[0], 0)

# Show the plot
plt.show()

```
